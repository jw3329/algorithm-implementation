```python
def mergeSort(arr):
    return mergeSortHelper(arr,0,len(arr)-1)
```

```python
def mergeSortHelper(arr,left,right):
    if left == right: return arr
    midIndex = (right - left) // 2 + left
    leftSorted = mergeSortHelper(arr,left,midIndex)
    rightSorted = mergeSortHelper(arr,midIndex+1,right)
    newArr = arr.copy()
    i,j = left,midIndex+1
    index = left
    while i < midIndex+1 and j < right+1:
        if leftSorted[i] < rightSorted[j]:
            newArr[index] = leftSorted[i]
            i += 1
        else:
            newArr[index] = rightSorted[j]
            j += 1
        index += 1
    while i < midIndex+1:
        newArr[index] = leftSorted[i]
        i += 1
        index += 1
    while j < right+1:
        newArr[index] = rightSorted[j]
        j += 1
        index += 1
    return newArr
```

```python
import random
```

```python
n = 1000
randomArr = [random.randint(0,n) for _ in range(n)]
# randomArr = [4,3,2,1]
print(randomArr)
```

```python
print(mergeSort(randomArr))
```

```python
print(sorted(randomArr))
```

```python
print(mergeSort(randomArr) == sorted(randomArr))
```
